```python
import os
import numpy as np
from six.moves import cPickle
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
output_dir = '../results/chipseq_sweep'
expt_name = 'CTCF'
downsamples = np.array([500, 1000, 2500, 5000, 7500, 10000])

aug = []
finetune = []
standard = []
for downsample in downsamples:
    with open(os.path.join(output_dir, expt_name+'_downsample_'+str(downsample)+'.pickle'), 'rb') as f:
        results = cPickle.load(f)
    aug.append(results['aug']['auroc'])
    finetune.append(results['finetune']['auroc'])
    standard.append(results['standard']['auroc'])
    
    
N = len(downsamples)
fontsize=14


cmap = cm.ScalarMappable(cmap='tab10')
rgb = cmap.to_rgba(np.linspace(0, 1, 8))  

fig = plt.figure(figsize=(4,3.5))
plt.scatter(downsamples, np.mean(np.array(aug), axis=1), color=rgb[0,:], s=50)
plt.scatter(downsamples, np.mean(np.array(finetune), axis=1), color=rgb[0,:], s=50)
plt.scatter(downsamples, np.mean(np.array(standard), axis=1), color=rgb[0,:], s=50)


mean_vals = []
std_vals = []
for i in range(len(aug)):
    mean_vals.append(np.mean(aug[i]))
    std_vals.append(np.std(aug[i]))
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)
plt.plot(downsamples, mean_vals, color=rgb[0,:], alpha=1)
plt.fill_between(downsamples, mean_vals-std_vals, mean_vals+std_vals,
    alpha=0.3, edgecolor=rgb[0,:], facecolor=rgb[0,:])
plt.scatter(downsamples, mean_vals, color=rgb[0,:], s=50)

mean_vals = []
std_vals = []
for i in range(len(finetune)):
    mean_vals.append(np.mean(finetune[i]))
    std_vals.append(np.std(finetune[i]))
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)
plt.plot(downsamples, mean_vals, color=rgb[1,:], alpha=1)
plt.fill_between(downsamples, mean_vals-std_vals, mean_vals+std_vals,
    alpha=0.3, edgecolor=rgb[1,:], facecolor=rgb[1,:])
plt.scatter(downsamples, mean_vals, color=rgb[1,:], s=50)

mean_vals = []
std_vals = []
for i in range(len(standard)):
    mean_vals.append(np.mean(standard[i]))
    std_vals.append(np.std(standard[i]))
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)
plt.plot(downsamples, mean_vals, color=rgb[2,:], alpha=1)
plt.fill_between(downsamples, mean_vals-std_vals, mean_vals+std_vals,
    alpha=0.3, edgecolor=rgb[2,:], facecolor=rgb[2,:])
plt.scatter(downsamples, mean_vals, color=rgb[2,:], s=50)


ax = plt.gca();
ax.legend(['Augment', 'Finetune', 'Standard'], 
          loc='lower right', fontsize=fontsize, frameon=False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
    
plt.setp(ax.get_yticklabels(), fontsize=fontsize)
plt.setp(ax.get_xticklabels(), fontsize=fontsize)
plt.ylabel('Classification (AUROC)', fontsize=fontsize);
plt.xlabel('Training Set size', fontsize=fontsize);
plt.xticks([0, 2500, 5000, 7500, 10000])
plt.yticks([0.85, 0.9, 0.95, 1,])
fig.savefig(os.path.join(output_dir, 'data_size_sweep_CTCF.pdf'), format='pdf', dpi=200, bbox_inches='tight')
```
